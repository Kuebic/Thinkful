## Interpolate

````python
# Creating the dataframe  
df_source = pd.DataFrame({"Group":['G_01', 'G_01', 'G_01', 'G_02', 'G_02', 'G_02'], 
                   "A":[None, 2, 3, 100, 200, None], 
                   "B":[1, None, 3, 100, None, 300], 
                   "C":[1, 2, None, None, 200, 300]}) 

df = df_source.copy()
# Print the dataframe 
df
````

````python
print(df.shape)
df.isnull().sum()
````

````python
# to interpolate the missing values 
df.interpolate(method ='linear', limit_direction ='both') 
````

````python
print(df.shape)
df.isnull().sum()
````

````python
df
````

````python
df = df.interpolate(method ='linear', limit_direction ='both')

df
````
