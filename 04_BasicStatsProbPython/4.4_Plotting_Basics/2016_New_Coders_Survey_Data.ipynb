Put all imports up top.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

Read csv file.

```python
survey = pd.read_csv('2016-FCC-New-Coders-Survey-Data.csv')
#print(survey)
```

```python
#Filtering out all entries that do not have an age listed
survey_filter_out_nan_age = survey[survey['Age'] > 1.0]
age_df = (survey_filter_out_nan_age.loc[:, 'Age'])
age_range = int(max(age_df)-min(age_df))

plt.figure(figsize=(15, 5))

plt.hist(age_df, bins=age_range, color='red')
plt.title('Age Distribution of Participants')
plt.xlabel('Ages')
plt.ylabel('Number of Participants')

plt.show()
```

The above chart shows a histogram of the age of participants.  We see a peak around late 20's, indicating that around that age the bootcamp is most popular.

```python
age_grouped = survey_filter_out_nan_age.groupby('Age').aggregate(np.mean)
#print(age_grouped)

plt.figure(figsize=(15, 5))

plt.plot(age_grouped['BootcampFinish']*100, color='purple')
plt.xlabel('Age')
plt.ylabel('Percentage')
plt.title('Average Bootcamp Completion Based on Age')

plt.show()
```

In the graph above, we grouped those with similar age together and calculated what percentage of people of that age-group completed their camp.  Perhaps not the most significatns on the tail ends, due to minimal sampling sizes in those age ranges.  Perhaps there is nothing special to see here, but I was hoping to see if age has anything to do with the probability of completion.

```python
plt.figure(figsize=(15, 5))

plt.plot(age_grouped['BootcampLoanYesNo']*100, color='blue')
plt.xlabel('Age')
plt.ylabel('Percentage')
plt.title('Average Loan Acceptance Based on Age')

plt.show()
```

The above graph, everyone of the same ages are grouped together and calculated to see what percentage of their age-group used loans to attend the bootcamp.  Again, the tail ends of the charts are not helpful due to smaller sample sizes, but I was hoping to see if age was a factor if participants required a loan and if they were accepted for one.

```python
plt.figure(figsize=(15, 5))

plt.plot(age_grouped['BootcampPostSalary'], color='purple')
plt.xlabel('Age')
plt.ylabel('Salary Amount')
plt.title('Average Salary Post-Bootcamp Based on Age')

plt.show()
```

The above graph, everyone of the same ages are grouped together and calculated to see what their average salary were post-bootcamp.  Again, the tail ends of the charts are not helpful due to smaller sample sizes, but I was hoping to see if age was a factor in their starting salary once they were out of the bootcamp.

```python
survey_filter_out_nan_salary = survey[survey['BootcampPostSalary'] > 1.0]
salary_df = (survey_filter_out_nan_salary.loc[:, 'BootcampPostSalary'])

plt.figure(figsize=(15, 5))

plt.hist(salary_df, bins=20, color='green')
plt.title('Salary Distribution of Participants')
plt.xlabel('Salary')
plt.ylabel('Number of Participants')

plt.show()
```

In the above graph, I created a histogram of the starting salaries reported and to see what the most common salaries were.  Surprising number of them list below $20,000.

```python
survey_filter_out_nan_salary_and_age = survey_filter_out_nan_age[survey_filter_out_nan_age['BootcampPostSalary'] > 1.0]
sna = survey_filter_out_nan_salary_and_age

plt.figure(figsize=(15,5))

plt.scatter(
    x=sna['Age'],
    y=sna['BootcampPostSalary'],
    color='green',
    marker='x', s=10
)
plt.title('Salary Based on Age')
plt.xlabel('Age')
plt.ylabel('Salary Post-Bootcamp')

plt.show()
```

The above graph tries to answer the similar question as the ones above, but we've plotted every salary based on age to see the distribution from a different perspective.  Especially interesting are where they plots clump together.
